```python
import pandas as pd 
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.ticker import ScalarFormatter
```

```python
final_2000 = "Data/final_tmdb_data_2000.csv.gz"
df_2000 = pd.read_csv(final_2000)
df_2000.head()
```

```python
df_2000.info()
```

```python
final_2001 = "Data/final_tmdb_data_2001.csv.gz"
df_2001 = pd.read_csv(final_2001)
df_2001.head()
```

```python
df_2001.info()
```

```python
frames = [df_2000, df_2001]
new_df = pd.concat(frames)
new_df.info()
```

```python
new_df.to_csv('Data/final_results_tmdb.csv.gz', compression='gzip',index=False)
```

```python
new_df[["budget", "revenue"]] = new_df[["budget", "revenue"]].replace(0, np.nan)
new_df.head()
```

```python
budget_filter = new_df["budget"] > 0.0
revenue_filter = new_df["revenue"] > 0.0

financials = new_df.loc[budget_filter | revenue_filter, :]
financials.head()
```

```python
financials.info()
```

```python
fig, ax = plt.subplots(figsize=(10, 4))

ax = financials["certification"].value_counts().plot(kind = "bar")

plt.ylabel("Count", fontsize = 15, fontweight = "bold", labelpad = 20)
plt.xlabel("Rating", fontsize = 15, fontweight = "bold", labelpad = 20)

plt.xticks(rotation = 45);
plt.xticks(fontweight = "medium", fontsize = 15)

plt.yticks(fontweight = "medium", fontsize = 15)
plt.yticks(np.arange(0, 230, 20));
```

```python
revenue = financials.groupby("certification", as_index = False)['revenue'].mean()
revenue.plot(kind = "bar", x = "certification", y = "revenue")

plt.ticklabel_format(axis = "y")

plt.ylabel("Average Revenue", fontsize = 15, fontweight = "bold", labelpad = 20)
plt.xlabel("Rating", fontsize = 15, fontweight = "bold", labelpad = 20)

plt.xticks(rotation = 45);
plt.xticks(fontweight = "medium", fontsize = 15)

plt.yticks(fontweight = "medium", fontsize = 15);
```

```python
budget = financials.groupby("certification", as_index = False)["budget"].mean()
budget.plot(kind = "bar", x = "certification", y = "budget")

plt.ticklabel_format(axis = "y", style='plain')

plt.ylabel("Average Budget", fontsize = 15, fontweight = "bold", labelpad = 20)
plt.xlabel("Rating", fontsize = 15, fontweight = "bold", labelpad = 20)

plt.xticks(rotation = 45);
plt.xticks(fontweight = "medium", fontsize = 15)

plt.yticks(fontweight = "medium", fontsize = 15);
```
